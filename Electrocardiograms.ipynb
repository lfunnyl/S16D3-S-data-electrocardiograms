##  (3) Logistic Regression

ðŸŽ¯ GÃ¶reviniz **kardiyovaskÃ¼ler hastalÄ±k riski altÄ±ndaki kalp atÄ±ÅŸlarÄ±nÄ± iÅŸaretlemektir.**

ðŸ‘‡ Bu gÃ¶revde `LogisticRegression`'Ä±n performansÄ±nÄ± araÅŸtÄ±rarak baÅŸlayalÄ±m. Modeli aÅŸaÄŸÄ±daki metriklerde deÄŸerlendirmek iÃ§in ***Ã§apraz doÄŸrulama kullanÄ±n***:
- DoÄŸruluk (Accuracy)
- HatÄ±rlama (Recall)
- Kesinlik (Precision)
- F1

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate
X = data.drop('target',axis=1)
y= data['target']

model =LogisticRegression(max_iter=1000)

log_cv_result=cross_validate(model,
                             X,
                             y,
                             cv=10,
                             scoring=['accuracy','recall','precision','f1']
                            )
for metric in ['accuracy', 'recall', 'precision', 'f1']:
    scores = log_cv_result[f'test_{metric}']
    print(f"{metric.capitalize():<10}: {scores.mean():.3f} Â± {scores.std():.3f}")

```

â“ **Soru (DoÄŸru tahminlerin oranÄ±)** â“ 

Bu model iÃ§in doÄŸru tahminlerin oranÄ± nedir? CevabÄ±nÄ±zÄ± `correct_pred_ratio` deÄŸiÅŸken adÄ±yla kaydedin.

```python
correct_pred_ratio=round(log_cv_result['test_accuracy'].mean(),2)
correct_pred_ratio
```

â“ **Soru (Riskli hastalarÄ± iÅŸaretleme yeteneÄŸi)** â“ 

Model riskli kalp atÄ±ÅŸlarÄ±nÄ±n yÃ¼zde kaÃ§Ä±nÄ± iÅŸaretleyebiliyor? CevabÄ±nÄ±zÄ± `flag_ratio` deÄŸiÅŸken adÄ±yla kaydedin.

```python
flag_ratio=round(log_cv_result['test_recall'].mean(),2)
flag_ratio
```

â“ **Soru (DoÄŸru iÅŸaretleme yeteneÄŸi)** â“ 

Model riskli bir kalp atÄ±ÅŸÄ± sinyali verdiÄŸinde, ne sÄ±klÄ±kta doÄŸrudur? CevabÄ±nÄ±zÄ± `correct_detection_ratio` deÄŸiÅŸken adÄ±yla kaydedin.

```python
correct_detection_ratio= round(log_cv_result['test_precision'].mean(),2)
correct_detection_ratio
```

â“ **Soru (Ã‡ok fazla yanlÄ±ÅŸ alarm olmadan mÃ¼mkÃ¼n olduÄŸunca Ã§ok riskli hastayÄ± tespit etme)** â“ 

Modelin yanlÄ±ÅŸ alarmlarÄ± sÄ±nÄ±rlarken mÃ¼mkÃ¼n olduÄŸunca Ã§ok riskli kalp atÄ±ÅŸÄ±nÄ± iÅŸaretleme yeteneÄŸi nedir? CevabÄ±nÄ±zÄ± `aggregated_metric` deÄŸiÅŸken adÄ±yla kaydedin.

```python
aggregated_metric = round(log_cv_result['test_f1'].mean(),2)
aggregated_metric
```

ðŸ§ª **Kodunuzu kontrol edin**

```python
from nbresult import ChallengeResult

result = ChallengeResult('logistic_regression_evaluation',
                         accuracy = correct_pred_ratio,
                         recall = flag_ratio,
                         precision = correct_detection_ratio,
                         f1 = aggregated_metric)
result.write()
print(result.check())
```

## (4) KarÄ±ÅŸÄ±klÄ±k Matrisi

ðŸ‘‡ `ConfusionMatrixDisplay` ([dokÃ¼mantasyon](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.ConfusionMatrixDisplay.html#sklearn.metrics.ConfusionMatrixDisplay)) kullanarak, Logistic Regression modelinin tahmin dÃ¶kÃ¼mÃ¼nÃ¼ gÃ¶rselleÅŸtirin.

<details>
<summary>ðŸ’¡ Ä°puÃ§larÄ±</summary>

    
1. [from_estimator](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.ConfusionMatrixDisplay.html#sklearn.metrics.ConfusionMatrixDisplay.from_estimator)    
2. [from_predictions](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.ConfusionMatrixDisplay.html#sklearn.metrics.ConfusionMatrixDisplay.from_predictions)
    
- Veri setinizi [`train-test-split`](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html) ile ayÄ±rmak ve test setindeki karÄ±ÅŸÄ±klÄ±k matrisine bakmak iÃ§in **Holdout yÃ¶ntemine** geri dÃ¶nmeyi unutmayÄ±n.  
</details>

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

X_train, X_test,y_train,y_test=train_test_split(
    X,y,test_size=0.3,stratify=y, random_state=42
)
print(f"Train set boyutu: {X_train.shape}")
print(f"Test set boyutu: {X_test.shape}")
print(f"Train'de riskli oranÄ±: {y_train.sum()/len(y_train):.2%}")
print(f"Test'te riskli oranÄ±: {y_test.sum()/len(y_test):.2%}")
```

```python
from sklearn.linear_model import LogisticRegression

# Model oluÅŸtur
model = LogisticRegression(max_iter=1000, random_state=42)

# Sadece TRAIN verisiyle eÄŸit
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Confusion matrix'i Ã§iz
ConfusionMatrixDisplay.from_estimator(
    model,                           # EÄŸitilmiÅŸ model
    X_test,                          # Test Ã¶zellikleri
    y_test,                          # Test gerÃ§ek deÄŸerleri
    display_labels=['SaÄŸlÄ±klÄ±', 'Riskli'],  # Etiketler
    cmap='Blues',                    # Renk paleti
    colorbar=True                    # Renk Ã§ubuÄŸu
)

plt.title('Logistic Regression - Confusion Matrix')
plt.show()
```

```python
# Ã–nce tahmin yap
y_pred = model.predict(X_test)

# Sonra confusion matrix Ã§iz
ConfusionMatrixDisplay.from_predictions(
    y_test,                          # GerÃ§ek deÄŸerler
    y_pred,                          # Tahminler
    display_labels=['SaÄŸlÄ±klÄ±', 'Riskli'],
    cmap='Reds'
)

plt.title('Logistic Regression - Confusion Matrix (Method 2)')
plt.show()
```

â„¹ï¸ KarÄ±ÅŸÄ±klÄ±k matrisi, modelin sÄ±nÄ±f dengesizliÄŸinden etkilendiÄŸini gÃ¶stermelidir: kalp atÄ±ÅŸlarÄ±nÄ± Ã§oÄŸu zaman saÄŸlÄ±klÄ± olarak tahmin eder. Bu davranÄ±ÅŸ nedeniyle, model sÄ±klÄ±kla doÄŸrudur ve **yÃ¼ksek doÄŸruluÄŸa** sahiptir. Ancak bu aynÄ± zamanda birÃ§ok riskli kalp atÄ±ÅŸÄ±nÄ± kaÃ§Ä±rmasÄ±na da neden olur: **kÃ¶tÃ¼ hatÄ±rlama** oranÄ±na sahiptir...

ðŸ‘‰ Bu nedenle bu model **riskli gÃ¶zlemleri iÅŸaretleme** gÃ¶revinde zayÄ±ftÄ±r.

â—ï¸ DoÄŸrulukla kandÄ±rÄ±lmayÄ±n ve gÃ¶revinize karÅŸÄ±lÄ±k gelen metriÄŸe bakÄ±n! â—ï¸

## (5) KNN SÄ±nÄ±flandÄ±rÄ±cÄ±

â“ **Soru (SÄ±nÄ±flandÄ±rma Model SeÃ§imi)** â“ 

VarsayÄ±lan bir KNN sÄ±nÄ±flandÄ±rÄ±cÄ±, riskli gÃ¶zlemleri iÅŸaretleme gÃ¶revinde daha iyi performans gÃ¶sterir mi?

CevabÄ±nÄ±zÄ± "KNN" veya "LogisticRegression" olarak `best_model` altÄ±nda kaydedin.

```python
from sklearn.neighbors import KNeighborsClassifier
knn_cv_results = cross_validate(KNeighborsClassifier(n_neighbors=5),
                                X, y,
                                cv=10,
                                scoring=['recall'])
knn_score = knn_cv_results['test_recall'].mean()
print(knn_score)

best_model = "KNN"
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_validate

# KNN'i deÄŸerlendir
knn_cv_results = cross_validate(KNeighborsClassifier(n_neighbors=5),
                                X, y,
                                cv=10,
                                scoring=['recall'])
knn_score = knn_cv_results['test_recall'].mean()

# LogReg'i hatÄ±rla (Ã¶nceki bÃ¶lÃ¼mden)
logreg_score = flag_ratio  # veya tekrar hesapla

# KarÅŸÄ±laÅŸtÄ±r
if knn_score > logreg_score:
    best_model = "KNN"
    print(f"âœ… KNN kazandÄ±! (Recall: {knn_score:.2%} vs {logreg_score:.2%})")
else:
    best_model = "LogisticRegression"
    print(f"âœ… LogReg kazandÄ±! (Recall: {logreg_score:.2%} vs {knn_score:.2%})")
```

ðŸ’ª Bu ECG veri seti iÃ§in, KNN SÄ±nÄ±flandÄ±rÄ±cÄ± LogisticRegression'dan Ã§ok daha yÃ¼ksek hatÄ±rlama oranÄ±na sahip olmalÄ± ve bu nedenle gÃ¶rev iÃ§in daha uygun.

ðŸ§ª **Kodunuzu kontrol edin**

```python
from nbresult import ChallengeResult

result = ChallengeResult('best_model',
                         model = best_model)
result.write()
print(result.check())
```
